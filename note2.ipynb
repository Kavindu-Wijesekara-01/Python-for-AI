# Operators

```python
#Arithmetc operator

print(6+5)
print(6-5)
print(6*5)
print(4/2)
print(4//2)
print(8%2)
print(5**2)
```

```python
# Relational Oparators

print(4>5)
print(4<=5)
print(4==5)
print(4!=5)
```

```python
#logical operators

print(1 and 0)
print(1 or 0)
print(not 0)
```

```python
# Bitwise operators

print(2 & 3) #bitwise and
print(2 | 3) #bitwise or
print(2 ^ 3) #bitwise xor
print(~3) #bitwise not
print(4 >> 2) #right shift
print(5 << 2) #left shift
```

```python
# Assignment operators

a = 6 
a += 5
print(a)
```

```python
# Membership operators

  # in / not in (it's depend on data type)
 
print('S' in 'Sri lanka')
print('S' not in 'sri lanka')
print(1 in [1,5,8,6])
```
