# Parameters Vs Arguments

```python
def odd_even(num): #Parameters
    """
    This function checks if a given number is odd or even.
    input: any valid number/integer
    output: odd / even
    Created Date: 8-14-2024
    """
    if type(num) == int:
        if num % 2 == 0:
            return "Even Number"
        else:
            return "Odd Number"
    else:
        return "Wrong input"
```

```python
res = odd_even(5) #Arguments
print(res)
```

# Types of Arguments

- Default Argument
- Positional Argument
- Keyword Argument

```python
def power(a=2,b=3):
    return a**b
```

```python
power()
```

```python
def power(a=2,b=3):
    return a**b
```

```python
power(2,4)
```

```python
# Positional Argument

def power(a=2,b=3):
    return a**b
```

```python
power(3,2)
```

```python
#Keyword Argument
def power(a=2,b=3):
    return a**b
```

```python
power(b=2,a=3)
```
