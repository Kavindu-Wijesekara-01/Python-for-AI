### List membership operations

```python
l1 = [1,2,3,4,5]
l2 = [6,7,8,9]
print(2 in l1)
print(3 in l2)
```

```python
for i in l1:
    print(i, end=" ")
```

```python
l = []
for i in range(1,11):
    l.append(i)
print(l)
```

```python
l = [i for i in range(1,11)]
print(l)
```

```python
v = [2,3,4]
s = -3
l = []

for i in v:
    result = i*s
    l.append(result)
print(l)
```

```python
l = [i*s for i in v]
print(l)
```

```python
l = [8,6,5,7]
lc = [i**2 for i in l]
print(lc)
```

```python
for i in range(1,51):
    if i % 5 == 0:
        print(i)
```

```python
l = [i for i in range(1,51) if i%5 ==0]
print(l)
```

```python
# find language which start with letter pass
languages = ["java","python","php","c","javascript"]

lc = [language for language in languages if language.startswith("p")]
print(lc)
```

```python
car = ["audi","bmw","tata","benz"]
no = [50,36,37,65]

for i,j in zip(car,no):
    print(f"{i} = {j}")
```

```python
lc = [f"{i} = {j}" for i,j in zip(car,no)]
```

```python
l = [2,4,6,10,1]
result=[]
for i in l:
    sum = 0
    for j in l:
        if i<= j:
            sum=sum+j
    result.append(sum)
print(result)
```

```python
num1 = [23,55,54,74,57,27]
num2 = [53,74,67,27,98,77]
common = []

for i in num1:
    if i in num2:
        if i not in common:
            common.append(i)
print(common)
```
