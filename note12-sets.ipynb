```python
s1 = {1,2,5,4,7}
s2 = {9,8,3,6}

s1.union(s2)
```

```python
s1.update(s2)
print(s1)
print(s2)
```

```python
s1 = {1,2,5,4,7}
s2 = {9,8,3,6,4,2}

s1.intersection_update(s2)
print(s1)
print(s2)
```

```python

s1 = {1,2,5,4,7}
s2 = {9,8,3,6,4,2}
s1.difference(s2)
```

```python
s1.difference_update(s2)
print(s1)
print(s2)
```

```python
s1 = {1,2,3,4}
s2 = {7,8,5,6}

s1.isdisjoint(s2)
```

```python
s1 = {1,2,3,4,5}
s2 = {3,4,5}

s1.issuperset(s2)
```

```python
s1 = {1,2,3}
s2 = s1.copy()

print(s1)
print(s2)
```

```python
fs1 = frozenset([1,2,3,4])
fs2 = frozenset([3,4,5,6])
print(fs1)
print(fs2)
```

```python
print(type(fs1))
```

```python
fs1 | fs2
```

```python
fs1 & fs2
```
