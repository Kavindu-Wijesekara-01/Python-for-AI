## While loop

```python
number = int(input("enter a number: "))

i = 1

while i<11:
    print(i*number, end=" ")
    i += 1
else:
    print("\n")
    print("programe exicuted")
```

```python
import random

jackport = random.randint(1,100)
guss_number = int(input("input a number: "))

while guss_number != jackport:
    if guss_number < jackport:
        print("wrong, enter higher number")
    else:
        print("wrong, enter lower number")
    guss_number = int(input("input a number: "))

else:
    print("correct guess")
```

## For loop

```python
for i in range(1,10):
    print(i)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(1,10):
    print("kavindu")
```

```python
for i in "kavindu":
    print(i, end="")
```

```python
for i in [1,25,36,"kavindu"]:
    print(i)
```

```python
for i in (1,25,36):
    print(i)
```

```python
for i in range(10,0,-1):
    print(i)
```

### Nested Loop

```python
row = int(input("input row number: "))

for i in range(1,row+1):
    for j in range(1,i+1):
        print("*", end="")
    print("")
```

```python
name = ["kavindu","tharushi","nethmi"]
for i in name:
    if i == "nethmi":
    
        break
    print(i)
```

```python
lower = int(input("enter lower: "))
upper = int(input("enter upper: "))

for i in range(lower,upper+1):
    for j in range(2,i):
        if i%j == 0:
            break
        else:
            continue
    else:
        print(i)
```
